```python
x = df.drop('House_Price',axis=1)
print(x)
y = df['House_Price']
print(y)
```

```python
# splitting the x and y values as train and test --- using train_test_split()
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### LINEAR REGRESSION

```python
#Model implementation

from sklearn.linear_model import LinearRegression
# LinearRegression is a class , to use a class first we have to create object

linear_regressor = LinearRegression()
#whatever the object name we give it is model(linear_regressor)

#training model
linear_regressor.fit(x_train,y_train)
```

```python
#testing model (predicting results)

y_pred_lr = linear_regressor.predict(x_test)
y_pred_lr
```

```python
#Model evaluation 
#checking results based on y_test
print("R^2 score of LR on training set:",linear_regressor.score(x_train,y_train))
print("R^2 score of LR on test set:",linear_regressor.score(x_test,y_test))
```

```python
#checking for error (checking wrong prediction)
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse

print("---Linear Regression-Model evaluation---")
print("Mean Absolute Error(MAE):",mae(y_test,y_pred_lr))
print("Mean Squared Error(MSE):",mse(y_test,y_pred_lr))
print("Root Mean Squared Error(RMSE):",np.sqrt(mse(y_test,y_pred_lr)))
```

```python
# decision tree
from sklearn.tree import DecisionTreeRegressor
dt_regressor = DecisionTreeRegressor()
dt_regressor.fit(x_train,y_train)
```

```python
#testing model (predicting results)

y_pred_dt = dt_regressor.predict(x_test)
y_pred_dt
```

```python
#Model evaluation 
#checking results based on y_test
print("R^2 score of DT on training set:",dt_regressor.score(x_train,y_train))
print("R^2 score of DT on test set:",dt_regressor.score(x_test,y_test))
```

```python
#checking for error (checking wrong prediction)
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse

print("---Decicion Tree-Model evaluation---")
print("Mean Absolute Error(MAE):",mae(y_test,y_pred_dt))
print("Mean Squared Error(MSE):",mse(y_test,y_pred_dt))
print("Root Mean Squared Error(RMSE):",np.sqrt(mse(y_test,y_pred_dt)))
```

```python
# Random Forest
from sklearn.ensemble import RandomForestRegressor
random_regressor = RandomForestRegressor()
random_regressor.fit(x_train,y_train)
```

```python
#testing model (predicting results)

y_pred_rf = random_regressor.predict(x_test)
y_pred_rf
```

```python
#Model evaluation 
#checking results based on y_test
print("R^2 score of DT on training set:",random_regressor.score(x_train,y_train))
print("R^2 score of DT on test set:",random_regressor.score(x_test,y_test))
```

```python
#checking for error (checking wrong prediction)
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse

print("---Decicion Tree-Model evaluation---")
print("Mean Absolute Error(MAE):",mae(y_test,y_pred_rf))
print("Mean Squared Error(MSE):",mse(y_test,y_pred_rf))
print("Root Mean Squared Error(RMSE):",np.sqrt(mse(y_test,y_pred_rf)))
```
